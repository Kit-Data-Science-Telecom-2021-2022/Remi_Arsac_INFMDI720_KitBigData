Ordre et noms des colonnes : 'year', 'name', 'gender', 'births'

Le DataFrame doit être trié selon l'année croissante puis selon l'ordre défini par les différents fichiers (voir la documentation du dataset).

```python
# names us
def df_names_us():
    df = pd.DataFrame()
            
    for year in range(1880,2021):
        names_us_of_the_year = pd.read_csv('./names/yob'+str(year)+'.txt', names=['name', 'gender', 'births'])
        names_us_of_the_year['year'] = year
        names_us_of_the_year = names_us_of_the_year[['year', 'name', 'gender', 'births']]
        
        df = df.append(names_us_of_the_year)
    
    df = df.reset_index(drop=True)
    
    return df
```

## 2. Prénoms français

On va s'intéresser au dataset **Fichiers France hors Mayotte** de l'INSEE :  https://www.insee.fr/fr/statistiques/2540004/

L'idée est de charger les données et ensuite de les conformer au DataFrame des prénoms US. Ainsi, toute manipulation sur le DataFrame des prénoms US pourra être directement réutilisée avec le DataFrame des prénoms français.
 
1. Télécharger le dataset des prénoms français : https://www.insee.fr/fr/statistiques/fichier/2540004/nat2020_csv.zip


Lire la documentation, ça peut être utile...
 
2. Implémenter une fonction Python qui produit un DataFrame avec les prénoms français en prenant le DataFrame des prénoms US comme modèle :
 
 - Même ordre et mêmes noms des colonnes : year, name, gender, births
 - Mêmes dtypes pour les colonnes
 - Mêmes valeurs pour la colonne 'gender'
 - Seuls les prénoms de 2 caractères et plus sont conservés
 - La casse des prénoms doit être identique : initiales en majuscule, autres lettres en minuscule
 - Les lignes avec des données inutilisables doivent être supprimées
 - Les données sont triées à l'identique : années (↑), puis gender (↑), puis births (↓) et enfin name (↑)
 - L'index du DataFrame doit aller de 0 à N-1

```python
# names fr
def df_names_fr():
    df = pd.DataFrame()
            
    df = pd.read_csv('./nat2020_csv/nat2020.csv', sep=';')
    
    #l0 = len(df.index)        
    #print('Step 0 len(df.index) l0 = '+str(l0))
    
    # Même ordre et mêmes noms des colonnes : year, name, gender, births
    df = df.rename(columns={'sexe':'gender', 'preusuel':'name', 'annais':'year', 'nombre':'births'})
    df = df[['year', 'name', 'gender', 'births']]
    
    #l1 = len(df.index)    
    #print('Step 1 len(df.index) l1 = '+str(l1)+' / l0-l1 = '+str(l0-l1))
    
    # Mêmes valeurs pour la colonne 'gender'
    mapping_gender = {1: 'M', 2: 'F'}
    df['gender'] = df['gender'].map(mapping_gender)
    
    #l2 = len(df.index) 
    #print('Step 2 len(df.index) l2 = '+str(l2)+' / l1-l2 = '+str(l1-l2))
    
    # Les lignes avec des données inutilisables doivent être supprimées
    # Suppression des lignes avec year='XXXX'
    #df_year_XXXX = df.loc[df['year']=='XXXX']       
    #print('Step 3.0 len(df_year_XXXX.index) = '+str(len(df_year_XXXX.index)))    
    
    df = df.loc[df['year']!='XXXX']
    
    #l3 = len(df.index)
    #print('Step 3.1 len(df.index) l3 = '+str(l3)+' / l2-l3 = '+str(l2-l3))
        
    # Suppression des lignes avec name='_PRENOMS_RARES'
    #df_PRENOMS_RARES = df.loc[df['name']=='_PRENOMS_RARES']
    #print('Step 4.0 len(df_PRENOMS_RARES.index) = '+str(len(df_PRENOMS_RARES.index)))
    
    df = df.loc[df['name']!='_PRENOMS_RARES']
        
    #l4 = len(df.index)
    #print('Step 4.1 len(df.index) l4 = '+str(l4)+' / l3-l4 = '+str(l3-l4))
    
    # Mêmes dtypes pour les colonnes
    # Conversion de la colonne year de type object en int64
    df['year'] = df['year'].astype(str).astype(int)
    
    #l5 = len(df.index)
    #print('Step 5 len(df.index) l5 = '+str(l5)+' / l4-l5 = '+str(l4-l5))
    
    # Seuls les prénoms de 2 caractères et plus sont conservés
    df['name'] = df['name'].astype(str)
    #df_prenoms_1_car = df.loc[df['name'].str.len() < 2]
    #print('Step 6.0 len(df_prenoms_1_car.index) = '+str(len(df_prenoms_1_car.index)))
    
    df = df.loc[df['name'].str.len() >= 2]
    
    #l6 = len(df.index)
    #print('Step 6.1 len(df.index) l6 = '+str(l6)+' / l5-l6 = '+str(l5-l6))
    
    # La casse des prénoms doit être identique : initiales en majuscule, autres lettres en minuscule
    df['name'] = df['name'].str.title()
    
    #l7 = len(df.index)
    #print('Step 7 len(df.index) l7 = '+str(l7)+' / l6-l7 = '+str(l6-l7))
    
    # Les données sont triées à l'identique : années (↑), puis gender (↑), puis births (↓) et enfin name (↑)
    df = df.sort_values(['year', 'gender', 'births', 'name'], ascending = (True, True, False, True))
    #l8 = len(df.index)
    #print('Step 8 len(df.index) l8 = '+str(l8)+' / l7-l8 = '+str(l7-l8))
    
    # L'index du DataFrame doit aller de 0 à N-1
    df = df.reset_index(drop=True)
    #l9 = len(df.index)
    #print('Step 9 len(df.index) l9 = '+str(l9)+' / l8-l9 = '+str(l8-l9))
    
    return df
```

## 3. Taux de change

On va s'intéresser au dataset des cours des devises de la Banque de France :  http://webstat.banque-france.fr/fr/#/downloading

L'idée est de charger les données, de les nettoyer et de pouvoir accéder aux cours de certaines devises à partir de leur code ISO3.
 
1. Utiliser le dataset des taux de change fourni.


2. Implémenter une fonction qui produit un DataFrame avec les taux de change par date pour une liste de codes ISO3 de devises passée en argument. L'index du DataFrame doit correspondre aux dates (voir la fonction pd.to_datetime() avec le format '%d/%m/%Y'). Les colonnes du DataFrame doivent correspondre aux devises.

```python
import re
df_source = pd.read_csv('./Webstat_Export_20210929.csv', sep=';', header=None)
df_source = df_source.drop(range(0,2))
df_source = df_source.drop(range(3,6))
df_source.columns = df_source.iloc[0]
df_source = df_source.drop(2)
df_source = df_source.reset_index(drop=True)
df_source = df_source.rename(columns={'Unité :':'Date'})
df_source['Date'] = pd.to_datetime(df_source['Date'])
df_source = df_source.set_index('Date')
df_source
```

```python
def extract_cur_code(currency):
    match = re.search(r'\(.*\)', currency)
    return currency[match.start()+1:match.end()-1]
```

```python
for i in range(0,len(df_source.columns)):
    df_source = df_source.rename(columns={df_source.columns[i]:extract_cur_code(df_source.columns[i])})

df_source = df_source.rename(columns={'RUB) (RUB':'RUB'})
df_source
```

```python
df_source.info()
```

```python
# taux de change
def df_taux_change(devises):
    #df = pd.DataFrame([], columns=devises)
    df = df_source.copy()
    df = df[devises]
    
    # Suppression des lignes comportant des nan ou des '-'
    df.dropna(inplace=True)
    for i in range(len(devises)):    
        df = df.loc[df[devises[0]]!='-']
    
    # On s'assure que tous les taux de change sont bien des flotants
    for i, col in enumerate(devises):
        df[col] = df[col].replace(',','.', regex=True).astype(float)
    
    return df
```

### Tests

```python
import unittest

class Lesson4Tests(unittest.TestCase):
    def test_df_names_us(self):
        df = df_names_us()
        # colonnes
        self.assertEqual(list(df.columns), ['year', 'name', 'gender', 'births'])
        # lignes
        self.assertEqual(len(df), 2020863)
        # index
        self.assertTrue(isinstance(df.index, pd.core.indexes.range.RangeIndex))
        # test NaN
        self.assertTrue(df.loc[df.isnull().any(axis=1)].empty)
        
    def test_df_names_fr(self):
        df = df_names_fr()
        # colonnes
        self.assertEqual(list(df.columns), ['year', 'name', 'gender', 'births'])
        # lignes
        self.assertEqual(len(df), 630407)
        # index
        self.assertTrue(isinstance(df.index, pd.core.indexes.range.RangeIndex))
        # test names
        self.assertTrue(df.loc[df['name'].str.contains('^[A-Z]+(?:-[A-Z]+)?$')].empty)
        # test gender
        self.assertEqual(len(df), len(df.loc[df['gender']=='F']) + len(df.loc[df['gender']=='M']))
        # test NaN
        self.assertTrue(df.loc[df.isnull().any(axis=1)].empty)

    def test_df_taux_change(self):
        df = df_taux_change(['CHF', 'GBP', 'USD'])
        # colonnes
        self.assertEqual(list(df.columns), ['CHF', 'GBP', 'USD'])
        # index
        self.assertTrue(isinstance(df.index, pd.core.indexes.datetimes.DatetimeIndex))
        # types taux
        self.assertTrue((df.dtypes == 'float').all())
        # test NaN
        self.assertTrue(df.loc[df.isnull().any(axis=1)].empty)
```

```python
# run tests
def run_tests():
    test_suite = unittest.makeSuite(Lesson4Tests)
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(test_suite)
```

```python
# run tests

run_tests()
```
